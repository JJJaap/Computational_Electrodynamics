```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

x = np.linspace(0,1000,1000)

y = np.sin(x/2/np.pi)

yff = np.fft.fftshift(np.fft.fft(np.fft.fftshift(y)))

plt.plot(y)
plt.show()
plt.plot(x-500,np.abs(yff)**2)
plt.xlim()
plt.show()
```
